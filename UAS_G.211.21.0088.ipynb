```python
#Import Library, disini Saya menggunakan library Matplotlib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load Dataset
data = pd.read_csv("CurahHujan.csv")
```

```python
# Melihat dimensi data
data.shape
```

```python
# Menampilkan data 10 Teratas
data.head(12)
```

```python
# Menggunakan nama variabel yang berbeda
a = data['Bulan']
b1 = data['1981 - 1990']
b2 = data['1991 - 2000']
b3 = data['2001 - 2010']

# Mengubah ukuran kolom dan jarak antar kolom
width = 0.15
spacing = 0
```

```python
# Menentukan ukuran figure
plt.figure(figsize=(13, 5))

# Menghitung posisi grup berdasarkan lebar dan jarak antara grup
ticks = np.arange(len(a))
pos1 = ticks - (width + spacing)
pos2 = ticks
pos3 = ticks + (width + spacing)

# Variabel untuk menyimpan bar plot
bars1 = plt.bar(pos1, b1, width=width, align='center', label='1981 - 1990', color='blue')
bars2 = plt.bar(pos2, b2, width=width, align='center', label='1991 - 2000', color='red')
bars3 = plt.bar(pos3, b3, width=width, align='center', label='2001 - 2010', color='green')

# Memberikan label pada sumbu x dan y
plt.xlabel('Bulan', fontweight='bold')
plt.ylabel('Curah Hujan (mm)', fontweight='bold')

# Mengatur tick label pada sumbu x
plt.xticks(ticks, a)

# Menampilkan legenda
handles, labels = plt.gca().get_legend_handles_labels()
plt.legend(handles, labels, loc='upper center', bbox_to_anchor=(0.5, 1.08), ncol=3)

# Menampilkan grid
plt.grid(True, axis='y', linewidth='0.2', color='gray', zorder=0)

# Menampilkan grafik
plt.show()
```

```python
#Import Library
import datetime
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
# Mengimport datetime untuk mengatur format waktu
from datetime import datetime
```

```python
#menggabungkan data
data['datetime'] = data.apply(lambda row: datetime(int(row['tahun']), int(row['bulan']), int(row['tanggal']), int(row['jam'])), axis=1)

# Mengambil kolom untuk dijadikan chart
datetime_dt = data['datetime']
asimilasi = data['asimilasi']
observasi = data['observasi']
non_asimilasi = data['non asimilasi']

```

```python
# Definisikan ukuran figure
plt.figure(figsize=(13, 5))

# Membuat line chart dengan style ketebalan, marker, dan warna
plt.plot(datetime_dt, asimilasi, color='blue', linewidth=2, marker='o', label='Asimilasi')
plt.plot(datetime_dt, observasi, color='yellow', linewidth=2, marker='o', label='Observasi')
plt.plot(datetime_dt, non_asimilasi, color='red', linewidth=2, marker='o', label='Non-Asimilasi')

# Pemberian keterangan
plt.xlabel('Waktu')
plt.ylabel('Celcius')
plt.title('Suhu Permukaan')
plt.legend(loc="lower right")

for d in [20,21, 22, 23, 24, 25, 26, 27]:
  plt.axhline(y=d, color='gray', linestyle='-', linewidth=0.3)

plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%H:%M UTC'))
plt.gca().xaxis.set_major_locator(mdates.AutoDateLocator())

#Menampilkan Hasil
plt.show()
```
